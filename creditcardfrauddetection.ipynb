Under Sampling
Build a sample dataset containing similar distribution of normal and fraud

```python
legit_sample=legit.sample(n=155)
```

Conatenation two dataframes

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into features and Targets

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split The Data into Training and Testing Data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(Y.shape,Y_train.shape,Y_test.shape)
```

Model Training


Logistic Regression Model

```python
model=LogisticRegression()
```

```python
#training the Logistic Regression with Training Data
model.fit(X_train,Y_train)
```

Accuracy Score

```python
# accuracy on training data
X_train_predict=model.predict(X_train)
train_accu=accuracy_score(X_train_predict,Y_train)
```

```python
print("Acuuracy on Training Data: ",train_accu)
```

```python
#accuracy on test data
X_test_pred=model.predict(X_test)
test_acc=accuracy_score(X_test_pred,Y_test)
```

```python
print(test_acc)
```
